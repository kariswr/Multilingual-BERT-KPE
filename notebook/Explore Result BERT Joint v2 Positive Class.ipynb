```python
def load_json(data_type, description):
  path = 'dataset/SQuAD/v2.0/' + data_type + '_' + description + '.json'
  print('Loading file at', path)
  # with open(path, 'r') as fh:
  #   jd = json.load(fh)
  return pd.read_json(path)

def save_to_json(df, data_type, description):
  path = 'dataset/SQuAD/v2.0/' + data_type + '_' + description + '.json'
  df.to_json(path)
  print('Saving file to', path)

def save_list_to_txt(path, list_obj):
  if not os.path.exists(path):
    with open(path, 'w') as f:
      f.writelines("%s\n" % element.replace("\n", " ") for element in list_obj)
    print(f'Save to {path}')
  else :
    print(f'Cancel saving, Path already exist : {path}')

def load_list_from_txt(path):
  print("loading file at", path)
  mainlist = []
  infile = open(path,'r')
  for line in infile:
    mainlist.append(line)
  infile.close()
  return mainlist
  
def load_list_from_json(path):
  print("loading file at", path)
  mainlist = []
  infile = open(path,'r')
  for line in infile:
    mainlist.append(json.loads(line))
  infile.close()
  return mainlist

def save_listDict_to_txt(path, list_obj):
  if os.path.exists(path):
    with open(path, 'w') as f:
      f.write(json.dumps(list_obj))
      f.write("\n")
    print(f'Save to {path}')
  else :
    print(f'Cancel saving, Path already exist : {path}')

def load_listDict_from_txt(path):
  print("loading file at", path)
  mainlist = []
  infile = open(path,'r')
  for line in infile:
    mainlist.append(json.loads(line))
  infile.close()
  return mainlist[0]
```

```python
from fuzzywuzzy import fuzz
import re

WORD_SIMILARITY_THRESHOLD = 100

def get_answer(answer, dict_list):
  for element in dict_list:
    temp = re.sub(r'[_<>,\(\)\.\'%]', ' \g<0> ', element['text'].lower())
    temp = re.sub(r'\(.*?\)', '', temp)
    temp = re.sub(r'\[.*?\]', '', temp)
    temp = re.sub(r'\{.*?\}', '', temp)
    temp = temp.replace('-', ' ')
    temp = re.split(r'[^a-zA-Z0-9_<>,#&\+\*\(\)\.\'%]', temp)
    temp = ' '.join(temp)
    temp = re.sub(' {2,}|\n{1,}', ' ', temp)
    temp = temp.lstrip()
    temp = temp.rstrip()
    if fuzz.ratio(temp, answer) >= WORD_SIMILARITY_THRESHOLD:
      return element
  return None

def get_detailed_ordered_answer(ground_truth, df):
  ground_truth_detail = []
  count = 0
  for i in range(len(ground_truth)):
    temp = []
    for answer in ground_truth[i]:
      answer_detail = get_answer(answer, df['keyphrases'][i])
      if (answer_detail != None):
        answer_detail['text_bert'] = answer
        temp.append(answer_detail)
      elif ('digit' in answer):
        temp.append({'answer_start': -1, 'text': answer, 'text_bert': answer})
      else :
        count +=1
        print(i, answer, df['keyphrases'][i])
    temp = sorted(temp, key = lambda x: x['answer_start'])
    ground_truth_detail.append(temp)
  print('Not found', count)
  return ground_truth_detail
```

```python
import re
from unidecode import unidecode
import spacy

nlp = spacy.load("en_core_web_sm")

NONE_NER = '<none>'

def unicode_to_ascii(s):
    return unidecode(s)

def normalize_string(s, lower=False):
    s = unicode_to_ascii(s)
    if lower:
        s = s.lower()
    s = re.sub(' {2,}|\n{1,}', ' ', s)
    s = s.lstrip()
    s = s.rstrip()
    s = s.replace('`', '')
    return s

punctuations = '!"#$%&\'()*+-/;<=>@?[\\]^_`{|}~'
punctuations_regex = re.compile(r"([%s])" % punctuations)
real_separator_regex = re.compile(r"(([\.,:][^a-zA-Z0-9])|([\.,:]$))")
def tokenize(s):
    s = re.sub(punctuations_regex, r" \1 ", s)
    s = re.sub(real_separator_regex, r" \1", s)
    s = s.split()
    return s

def add_ner_ground_truth(ground_truth_answers, df):
  for i in range(len(ground_truth_answers)):
    # if (i == 1008):
    if (i != len(ground_truth_answers)):
      for j in range(len(ground_truth_answers[i])):
        pointer_loc = 0
        answer = ground_truth_answers[i][j]
        ground_truth_answers[i][j]['NER'] = []
        ground_truth_answers[i][j]['POS'] = []
        text = normalize_string(answer['text'])
        doc = nlp(text)
        answer_token = [token.text for token in doc]
        # answer_token = tokenize(normalize_string(answer['text']))
        ground_truth_answers[i][j]['token'] = answer_token.copy()
        if (answer['answer_start'] == -1):
          # print(i, j, "not found")
          ground_truth_answers[i][j]['NER'] = [NONE_NER] * len(answer_token)
          ground_truth_answers[i][j]['POS'] = [NONE_NER] * len(answer_token)
          continue
        for index_token in range(len(df['token'][i])):
          token = df['token'][i][index_token]
          pointer_loc += len(token) + 1
          # print(i,j, answer_token, token)
          if (answer['answer_start'] - pointer_loc <= 0):
            if (token == answer_token[0]):
              # print("masuk ini")
              ground_truth_answers[i][j]['NER'].append(df['NER'][i][index_token])
              ground_truth_answers[i][j]['POS'].append(df['POS'][i][index_token])
              answer_token.pop(0)
            elif (len(token) < len(answer_token[0]) and token == answer_token[0][:len(token)] and (answer_token[0] not in df['token'][i][index_token:])):
              # print("masuk itu")
              ground_truth_answers[i][j]['NER'].append(df['NER'][i][index_token])
              ground_truth_answers[i][j]['POS'].append(df['POS'][i][index_token])
              answer_token[0] = answer_token[0][len(token):]
            elif (len(token) > len(answer_token[0]) and (token[:len(answer_token[0])] == answer_token[0] or token[len(answer_token[0]):] == answer_token[0]) and (answer_token[0] not in df['token'][i][index_token:])):
              # print("masuk itu")
              ground_truth_answers[i][j]['NER'].append(df['NER'][i][index_token])
              ground_truth_answers[i][j]['POS'].append(df['POS'][i][index_token])
              answer_token.pop(0)
            # elif (len(token) < len(answer_token[0]) and token == answer_token[0][:len(token)])):
            #   print("masuk lain")
            #   ground_truth_answers[i][j]['NER'].append(df['NER'][i][index_token])
            #   answer_token[0] = answer_token[0][answer_token[0].find(token) + 1:]
          if not answer_token:
            break
      # if (i == 1008):
      #   break
  return ground_truth_answers
```

### Norm and Stem as implement in BERT-Joint

```python
import unicodedata
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

def norm_and_stem(phrase_list, merge=False):
    norm_stem_phrases = []
    for phrase in phrase_list:
        norm_chars = unicodedata.normalize('NFD', phrase)
        stem_chars = " ".join([stemmer.stem(w) for w in norm_chars.split(" ")])
        if merge:
            norm_stem_phrases.append(norm_chars)
            norm_stem_phrases.append(stem_chars)
        else:
            norm_stem_phrases.append((norm_chars, stem_chars))
    return norm_stem_phrases
```

```python
import collections

def sort_dict(dict_obj):
  dict_obj = sorted(dict_obj.items(), key=lambda x: x[1], reverse=True)
  dict_obj = collections.OrderedDict(dict_obj)
  return dict_obj

def get_ner_pos_distribution(predict_ner):
  ner_distribution = {}
  pos_distribution = {}
  for i in range(len(predict_ner)):
    for ner in predict_ner[i][1]:
      if ner in ner_distribution :
        ner_distribution[ner] += 1
      else :
        ner_distribution[ner] = 1
    for pos in predict_ner[i][2]:
      if pos in pos_distribution :
        pos_distribution[pos] += 1
      else :
        pos_distribution[pos] = 1
  ner_distribution = sort_dict(ner_distribution)
  pos_distribution = sort_dict(pos_distribution)
  return ner_distribution, pos_distribution
```
